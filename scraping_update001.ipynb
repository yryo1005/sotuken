```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
```

```python
driver = webdriver.Chrome()            # Chromeを準備
# 自分の行きたいサイトのurlを貼り付ける
url = 'https://store.steampowered.com/app/730/CounterStrike_2/'
# 自分の見たいサイトもurlを開く
driver.get(url)  

#By.XPATHでサイトの検証のタグからCopyでXPATHを出す
#elemにはアップデート情報のボタンのxpathを与えている
elem = driver.find_element(By.XPATH, '//*[@id="tabletGrid"]/div[1]/div[5]/div[2]/div[4]/div/div[2]/div[2]')

# 先ほど指定したクリックしたい要素を格納した変数elemにクリックを指示する
elem.click()
# 少し待機してクリックした後の遅延を待っている
time.sleep(1)

# ページのスクロール (レビューを読み込むため)
SCROLL_PAUSE_TIME = 1  # スクロール間の待機時間（秒）
datas = list()


# ブラウザのウインドウの高さを取得する
window_height = driver.execute_script("return window.innerHeight")

# スクロール開始位置
top_position = 1
page_num = 1
while 1:
    print(f"page{page_num}を読み込んでます。")
    # スクロール前のページの高さを取得
    last_height = driver.execute_script("return document.body.scrollHeight")

    top = top_position

    while top < last_height:        
        
        
        # 表示中のブラウザの9.5割スクロールする
        top += int(window_height * 0.80)
        driver.execute_script("window.scrollTo(0, %d)" % top)
        time.sleep(SCROLL_PAUSE_TIME)

    time.sleep(0.9)

    # スクロールし更新された後のブラウザの高さを取得
    new_last_height = driver.execute_script("return document.body.scrollHeight")

    # スクロール前後でのブラウザの高さに変化がなくなればループを終了する
    if last_height == new_last_height:
        break
    top_position = last_height

    # とりあえず page10くらいまで読み込む 以下の条件文を削除するとレビュー全部とれる気がする
    if page_num > 10:
        break

    page_num += 1

update_point = driver.find_element(By.CLASS_NAME, 'bbcodes_ListItem_3DlO0')

# while True:
# 投稿した日付を取得
dates = driver.find_elements(By.XPATH, '/html/body/div[3]/div[2]/div/div/div/div/div[2]/div[1]/div[1]/div/div[1]/div/div/div/div[2]')


# テキスト出力する
for date in dates:
    datas.append(date.text)

print(datas)

# time.sleep(5)
driver.quit()                          # ブラウザを閉じる
```
